```python
# Read the CSV file into a DataFrame
import pandas as pd
import io

# load your dataframe here
df = pd.read_csv(io.BytesIO(uploaded[r'5000_names_split1.csv'])) #change 'top1000names' to the file name after you upload it
```

```python
import requests
from bs4 import BeautifulSoup
#import pandas as pd

def get_h3_texts(div, num_elements=2):
    h3_tags = div.find_all('h3')
    h3_texts = []
    for i, h3 in enumerate(h3_tags):
        if i >= num_elements:
            break
        if h3.span:
            text = "".join(t for t in h3.contents if not hasattr(t, 'name') or t.name != 'span')
        else:
            text = h3.text
        h3_texts.append(text)
    return h3_texts

def fetch_webpage_and_extract_data(name, num_results=5):
    url = f'https://genealogy.org.il/AID/?queryGivenNameHE={name}'
    response = requests.post(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    results_divs = soup.find_all('div', class_="result clearfix")
    all_h3_texts = []
    for i, div in enumerate(results_divs):
        if i >= num_results:
            break
        h3_texts = get_h3_texts(div)
        print(h3_texts)  # prints current h3_texts
        all_h3_texts.append(h3_texts)
    return all_h3_texts

# Loop over all names
all_results = []
for name in df['CompanyName - Split 1']:
    name_results = fetch_webpage_and_extract_data(name)
    all_results.extend(name_results)

#make it a data frame with columns for each key
df_output = pd.DataFrame(all_results, columns=['English Name', 'Hebrew Name'])
df_output.to_csv('first_name_scraper_results.csv', index=False)
```
